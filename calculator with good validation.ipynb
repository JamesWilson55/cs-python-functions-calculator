```python
# function name: mainmenu().
# Purpose: to get the user input for what operation they want to do.
# Input: No input
# Returns: The capital letter from valid choices
def mainMenu():
    validChoices = ["A", "S","D","M","SQ","P"]
    valid = False
    while not valid:
        print("A)dd numbers")
        print("S)ubtract numbers")
        print("M)ultiply numbers")
        print("D)ivide numbers")
        print("SQ)ure number")
        print("P)ower")
        choice = input("choose your option").upper()
        if choice not in validChoices:
            print("sorry that is not a valid menu choice ")
        else:
            valid= True
            
    return choice
    
# function name: enterNumber().
# Purpose: to get the user input for what numbers they want to use.
# Input: No input
# Returns: The integer the user entered
def enterNumber():
    valid = False
    while not valid:
        try:
            x=int(input("enter the number - "))
            valid=True
        except ValueError:
            print("That was not a valid number.. please try again")
    
    return x

# function name: add().
# Purpose: adds the two numbers.
# Input: two integers
# Returns: the answer to the two numbers added
def add(x,y):
    
    return x+y

# function name: subtract().
# Purpose: subtracts the two numbers.
# Input: two integers
# Returns: the answer to the two numbers subtracted
def subtract(x,y):
    
    return x-y

# function name: multiply().
# Purpose: multiplies the two numbers.
# Input: two integers
# Returns: the answer to the two numbers multiplied
def multiply(x,y):
    
    return x*y

# function name: deivide().
# Purpose: divides the two numbers.
# Input: two integers
# Returns: the answer to the two numbers divided
def divide(x,y):
    
    return x/y

# function name: square().
# Purpose: squares the number.
# Input: one integer
# Returns: the answer to the number squared
def square(x,y):
    
    return x**2


#main program starts here
# purpose: uses the users option to send it to the correct function to do the right calculation.
print("welcome to my calculator")
again = "y"
while again == "y":
    
    Choice = mainMenu()
    if Choice == "A":
        answer = add(enterNumber(), enterNumber())
    elif Choice == "S":
        answer = subtract(enterNumber(), enterNumber())
    elif Choice == "M":
        answer = multiply(enterNumber(), enterNumber())
    elif Choice == "D":
        answer = divide(enterNumber(), enterNumber())
    elif Choice == "SQ":
        answer = square(enterNumber(),0)
       
    else:
        print("error - we should never get here")
    
    print("the answer from your calcaulation is = "+str(answer))
    
```
